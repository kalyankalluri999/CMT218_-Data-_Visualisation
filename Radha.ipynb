**Handling Missing Vaues**

*When information is lacking for one or more items, or for the entire unit, it is referred to as missing data. Missing data is a major issue in real-world situations. In Pandas, missing data is also referred to as NA (Not Available) values. Many datasets in DataFrame occasionally arrive with missing data, either because the data never existed or because it exists but was not collected. As an illustration, let's say that various survey participants decide not to disclose their income or address. As a result, numerous datasets would disappear.*


*Two values are used in Pandas to represent missing data:*

**None:** *In Python code, the singleton object None is frequently used to represent missing data.*

**Not a Number, or NaN,** *is a unique floating-point value that is supported by all systems that employ the IEEE standard floating-point representation*

*When it comes to indicating missing or null values, pandas treats None and NaN as practically equivalent. There are a number of helpful functions for finding, eliminating, and replacing null values in Pandas DataFrame to help with this convention:*

**isnull()**

**notnull()**

**dropna()**

**fillna()**

**replace()**

**interpolate()**

*As most of the have more tha 50% of null values hence dropping them*
